### **1.1/ WHY STUDY STATISTICS?**

- Since there exists variability in every possible aspect of this world, in order to better understand some subjects or statements/conclusions on certain problems, we need to somehow collect, organize and summarize the data available, from which realistic, data-driven opinions can be formed and adjust reasonably.
- The tools to do these tasks can be found in Statistics, Mathematics for analysis and in Technology for collection and computation.

### **1.2/ WHAT IS STATISTICS?**

- There are two types of statistics:
    + Descriptive Statistics: provides the means for data organization and summarization, allows people to deduct conclusions from facts.
    + Inferential Statistics: provides the means for generalization beyond actual data, allows people to predict/approximate future outcomes based on historical or a smaller set of data.

### **1.3/ MORE ABOUT INFERENTIAL STATISTICS**

#### **Populations and Samples:**
- Definition:
    + Population: any complete collection of actual observations or potential observations.
    + Sample: any smaller collection of actual observations from a population.
- Property:
    + How a data collection is considered a population or a sample is subjective to the observers, often decided by the range of subjects, on which the conclusions obtained from data are applied. For example, if one wants to indicate the average weight of students in a class, then the collection of weights of all students in that class is a population, however, if the average weight of all students in an university is the goal, then the avaiable data is only a sample.
- Notes:
    + Populations are usually large and exist only as potential observations, whereas samples are the actual observations. When samples are used as the basis for any implications on populations, it is important that the samples are good representatives of the populations, otherwise, the generalizaions might be erroneous.

#### **Random Sampling (Surveys):**
- Definition:
    + Random Sampling is a procedure designed to ensure that each potential observation in the population has an equal chance of being selected in a survey.
- Random Sampling is an important part of statistical examination, it should theoretically remove any obvious biases that might cause the sample to misrepresent the population. Nonetheless, the procedure can be quite difficult when a population:
    + Lacks structure or
    + Specific boundaries.
    in such cases, a random sample can only be approximated.

#### **Random Assignment (Experiments):**
- Definition:
    + Random Assignment is a procedure designed to ensure that each subject has an equal chance of being assigned to any group in an experiment.
- Random Assignment is applied when the observer does not have any control over the sampling procedure (dependent of volunteers) and comparison between samples is the goal of the experiment. It asserts that whether the differences between groups of samples are real or attributable to variability or chances.

#### **Overview: Surveys and Experiments**
- Based on random samples from populations, surveys permit generalizations from samples back to populations.
- Based on random assignment of volunteers to groups, experiments permit decisions abou whether differences between groups are real or merely transitory.

### **1.4/ THREE TYPES OF DATA**

- Any statistical analysis is performed on data - a collection of actual observations or scores in a survey or an experiment.
- The precise form of a statistical analysis often depends on whether data are qualitative, ranked, or quantitative.
- Definition:
    + Qualitative Data: a collection of observations where any single observation is a word, letter or any numerical code that represents a class or category.
    + Ranked Data: a collection of observations where any single observation is a number that indicates relative standing.
    + Quantitative Data: a collection of observations where any single observation is a number that represents an amount or a count.

### **1.5/ LEVELS OF MEASUREMENT**

- Definition:
    + Level of Measurement specifies the extent to which a number (or word or letter) actuall represents some attribute and, therefore, has implications for the appropriateness of various arithmetic operations and statistical procedures.
- The levels of measurement:
    + Nominal: for qualitative data.
    + Ordinal: for ranked data.
    + Interval/Ratio: for quantitative data.
- More complex levels of measurement are associated with numbers that, because they better represent attributes, permit a wider variety of arithmetic operations and statistical procedures.

#### **Qualitative Data and Nominal Measurement:**
- Definition:
    + Nominal Measurement: words, letters, or numerical codes of qualitative data that reflect differences in kind based on classification.
- Property:
    + Nomimal Measurement has only one single property, which is ***classification***. Since words, letters, or numerical codes only indicate the classes or categories, with which they belong, and there is no amount (in number) or level included.

#### **Ranked Data and Ordinal Measurement:**
- Definition:
    + Ordinal Measurement: relative standing of ranked data that reflects the differences in degree based on order.
- Property:
    + Ordinal Measurement's properties include ***classification***, since the data indicates the class or category that the observation belongs to, the difference is that the classes are placed in order.
    + A distinctive property of Ordinal Measurement is ***order***.
    + Even though ranked data often comes in the form of numbers, their characteristics cannot be conveyed simply by arithmetic operations because they do not just reflect an amount, but also the relative difference in degree between one another, and the differences between each class might not be the same.

#### **Quantitative Data and Interval/Ratio Measurement:**
- Definition:
    + Interval/Ratio Measurement: amounts or counts of quantitative data that reflect differences in degree based on **equal** intervals and a true zero.
- Property:
    + The distinctive properties of Interval/Ratio Measurement are **equal intervals and a true zero** (a *true zero* reflects the complete absence of any amount to be observed).
    + The number readings reflect exactly the amount that is intended to be calculated from an observation.

#### **Measurement of Nonphysical Characteristics:**
- When numbers represent nonphysical characteristics, the attainment of interval/ratio measurement of is questionable. This is because the intervals/ratios between these numbers might not absolutely reflect the amount of differences, instead, they might indicate the classes/categories with which data belong or a hierachy of sort so that observations can be ranked accordingly. Remember that the properties of Interval/Ratio Measurement includes a **true zero (absolute absence of any amount/count of an observation)** and **equal intervals**, without these, the results of arithmetic operations do not imply any conclusions.

### **1.6/ TYPES OF VARIABLES**

#### **General Definition:**
- A Variable: is a property or characteristics that can take on different values.

#### **Discrete and Continuous Variables:**
- Definition:
    + Discrete Variable: a variable that consists of isolated numbers, separated by gaps.
    + Continuous Variable: a variable that consists of numbers whose values, at least in theory, have no restrictions.

#### **Approximate Numbers:**
- Definition: numbers that are rounded off, as is always the case with values for continous variables.

#### **Independent and Dependent Variables:**
* Independent Variable:
    - Definition: the treatment manipulated by the investigator in an experiment (a variable that is in disregards of any affect occurred during the experiment, since its value was given before the experiment).
* Dependent Variable:
    - Definition: a variable that is believed to have been influenced by the independent variable (the record that is obtained after the experiment and only occurred due to the treatment selected).

#### **Observational Studies:**
- Definition: a study that focuses on detecting relationships between variables not manipulated by the investigator.

#### **Confounding Variable:**
- Definition: an uncontrolled variable that compromises the interpretation of an experiment.
