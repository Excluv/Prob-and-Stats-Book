### **3.1/ MODE**

- Definition:
    + Mode: the value of the most frequent score.

#### **More than One Mode:**
- As the typical shapes suggested in the previous chapter, one distribution might have more than one peak, which implies that there might be more than one mode in a distribution.
- The existence of multiple modes within one distribution reflects important differences among subsets of data.

**Progress Check 3.1: Determine the mode for the following retirement ages: 60, 63, 45, 63, 65, 70, 55, 63, 60, 65, 63.**

### **3.3/ MEAN**

- Definition:
    + Mean: the value found by dividing the sum of all scores by the total number of scores within a dataset.

#### **Sample or Population:**
- When the dataset is viewed:
    + As a sample of the population: then the Mean only serves as the Sample Mean, which returns the average value within the given dataset alone.
    + As a population: then the Mean is considered a Population Mean and is the Mean for all Samples.

#### **Formula for Sample Mean:**
- Formula: <br>
<center> $\LARGE\overline{X} = \frac{\sum{X}}{n}$ </center><br>
Where: <br>
   + $\overline{X}$: the Sample Mean. <br>
   + $\sum{X}$: the sum of all the scores within the dataset. <br>
   + n: the sample size (the total number of scores in the sample). <br>

#### **Formula for Population Mean:**
- Formula:
<center> $\LARGE\mu = \frac{\sum{X}}{N}$ </center>
Where: <br>
+ $\mu$: the Population Mean. <br>
+ $\sum{X}$: the sum of all the scores within the dataset.
+ N: the population size.

#### **Mean as Balance Point:**
- The Mean serves as the balance point for its frequency distribution, meaning:
    + The sum off scores, expressed as positive or negative deviations from the mean, always equals 0.
- Unlike the Mode and the Median, the Mean reflects all the values of a frequency distribution. That is also the factor that makes the Mean so sensitive to the change in scores, especially in the appearance of outliers.

### **3.4/ WHICH AVERAGE?**

#### **If Distribution is not Skewed:**
- When the distribution is not too skewed, any of the three "Average" can be used to describe the central tendency.

#### **If Distribution is Skewed:**
- If in a distribution:
    + The Mean exceeds the Median appriciably: the distribution is positively skewed and the outliers lie in the right of the ordered dataset.
    + The Median exceeds the Mean appriciably: the distribution is negatively skewed and the outliers lie in the left of the ordered dataset.

#### **Interpreting Differences between Mean and Median:**
- When the distribution is skewed, report both the Median and the Mean.

#### **3.18. Given that the mean equals 5, what must be the value of the one missing observation from each of the following sets of observations?** <br>
(a) 1, 2, 10. <br>
(b) 2, 4, 1, 5, 7, 7. <br>
(c) 6, 9, 2, 7, 1, 2. <br>

```python
# Solution:
def cal_dependent_var(data, mean):
    deviations = data - mean
    num = -deviations.sum() + 5
    return num
```

```python
    # a):
data = np.array([1, 2, 10])
num = cal_dependent_var(data, 5)
num
```

```python
    # b):
data = np.array([2, 4, 1, 5, 7, 7])
num = cal_dependent_var(data, 5)
num
```

```python
    # c):
data = np.array([6, 9, 2, 7, 1, 2])
num = cal_dependent_var(data, 5)
num
```
