### **18.1** A TWO-FACTOR EXPERIMENT: RESPONSIBILITY IN CROWDS

- Definition:
    + Two-Factor ANOVA: A more complex type of analysis that tests whether differences exist among population means catergorized by two factors or independent variables.

- Using the two-factor ANOVA, investigators can test not just two, but three null hypotheses on:
    + The effect of one factor (independent variable), called the factor #1, regardless of the other factor.
    + The effect of the other factor, called the factor #2, regardless of the first factor.
    + The effect of the combination or interaction between the two factors.

### **18.2** THREE F TESTS

- In two-factor ANOVA, three different null hypotheses are tested, one at a time, with three F ratios: $F_{column}, F_{row}, F_{interaction}$.
- The numerator of each of the three F ratios reflects a different aspect of variability between cells:
    + In $F_{column}$: the variability between columns.
    + In $F_{row}$: the variability between rows.
    + In $F_{interaction}$: any remaining variability that is not attributable to either variability between columns or between rows.
- The shaded numerator terms for the three F ratios reflect estimate random error and, if present, the treatment effect (for subjects treated differently). The denominator term always estimates only the random error (for subjects treated similarly in the same cell).

### **18.4** DETAILS: VARIANCE ESTIMATE

- Each of the three F ratios in a two-factor ANOVA is based on a ratio involving two variance estimates: a mean square in the numerator that reflects random error plus, if present, any specific treatment effect and a mean square in the denominator that reflects only random error.

#### **Sums of Squares (Definitional Formulas)**

<center><b>SUM OF SQUARES (TWO FACTOR)</b></center> <br>
<center>$\Large SS_{total} = SS_{column} + SS_{row} + SS_{interaction} + SS_{within}$</center>

- The computation of the various SS terms can be viewd as a two-step effort:

1. The two factors and their interactions are ignored, the total variability is partitioned into two essential components (as in any of the F tests for ANOVA): <br><br>
<center>$\large SS_{total} = SS_{between} + SS_{within}$</center> 
Where:
- $SS_{total} = \sum{(X - \overline{X}_{grand})^2}$ <br>
- $SS_{between} = n\sum{(\overline{X}_{cell} - \overline{X}_{grand})^2}$, where n, the sample size in each cell, adjusts for the fact that the deviation $\overline{X}_{cell} - \overline{X}_{grand}$ is the same for every score in its cell.
- $SS_{within/error} = \sum{(X - \overline{X}_{cell})^2}$.

2. Variability between cells, $SS_{between}$ is partitioned into three addtional sums of squares, $SS_{column}, SS_{row}, SS_{interaction}$ - that reflects the identifiable sources of treatment variability in the two-factor ANOVA: <br><br>
<center>$\large SS_{between} = SS_{column} + SS_{row} + SS_{interaction}$</center>
Where:
- The variability between columns $SS_{column} = rn\sum{(\overline{X}_{column} - \overline{X}_{grand})^2}$, where r is the number of rows and n is the sample size in each cell, and rn equals the sample size in each column. The product rn adjusts for the fact that the mean deviation, $\overline{X}_{column} - \overline{X}_{grand}$ is the same for every score in its column.
- The variability between rows $SS_{row} = cn\sum{(\overline{X}_{row} - \overline{X}_{grand})^2}$, where c is the number of columns and cn equals the sample size in each row.
- The remaining variability between cells, after the removal of column and cell variabilities $SS_{interaction} = SS_{between} - (SS_{column} + SS_{row})$.

<center>$\Large HSD = q\sqrt{\frac{MS_{within}}{n}}$</center>

Where:

    + n = c.n: the product of the number of columns and cell sample size, used if pairs of row means are being compared.

### **18.9** SIMPLE EFFECTS

- Whenver the interaction between independent variables is statistically significant, we can conduct a new $F_{se}$ test to identify the inconsistency among simple effects that produce the interaction.

- Definition:
    + $F_{se}$ Test for Simple Effects: A test of the effect of one factor on the dependent variable at a single level of another factor.

#### **$F_{se}$ Test for Simple Effects**

<center><b>$F_{se}$ RATIO (SIMPLE EFFECT)</b></center> <br>
<center>$\it \Large F_{se} = \frac{\text{MS}_{se}}{\text{MS}_{within}}$</center>

Where:
+ $MS_{se}$: the mean square for the variation of every cell mean in a single row (or a single column) about the overall or grand mean for the entire row (or column).
+ The degrees of freedom for the numerator of the single-effect $F_{se}$ ratio is the same as that for the corresponding main-effect $F$ ratio, namely $\it \text{df}_{between}$ (equals either r - 1 or c - 1).

#### **Calculating $F_{se}$**

<center><b>SUM OF SQUARES (SIMPLE EFFECT)</b></center> <br>
<center>$\Large SS_{se} = \sum{\frac{T_{se}^2}{n}} - \frac{G_{se}^2}{N_{se}}$</center>

Where:
+ $T_{se}$: the total of each cell in a single row (or column).
+ $G_{se}$: the total of an entire row (or column).
+ $n$: the cell sample size.
+ $N_{se}$: the sample size in an entire row (or column).

```python
def twofactorANOVA_graphs(data):
    fig, ax = plt.subplots(1, 3, figsize=(12, 3))
    fig.subplots_adjust(wspace=0.5)
    
    ticks_range = np.arange(np.max(data[0]["y"])) + 2
    for idx in range(2):
        ax[idx].plot(data[idx]["x"], data[idx]["y"], marker="o", color="black")
        ax[idx].set_title(data[idx]["title"])
        ax[idx].set_xlabel(data[idx]["xlabel"])
        ax[idx].set_ylabel(data[idx]["ylabel"])
        ax[idx].set_yticks(ticks_range)
    
    idx = 2
    ax[idx].plot(data[idx]["x"], data[idx]["y_1"], marker="o", linestyle="--", label=data[idx]["levels"][0])
    ax[idx].plot(data[idx]["x"], data[idx]["y_2"], marker="o", linestyle="-", label=data[idx]["levels"][1])
    try:
        ax[idx].plot(data[idx]["x"], data[idx]["y_3"], marker="o", linestyle="-.", label=data[idx]["levels"][2])
    except KeyError:
        pass

    ax[idx].set_title(data[idx]["title"])
    ax[idx].set_xlabel(data[idx]["xlabel"])
    ax[idx].set_ylabel(data[idx]["ylabel"])
    ax[idx].set_title(data[idx]["title"])
    ax[idx].legend(frameon=True, loc="upper left", bbox_to_anchor=(1.0, 1.0))
    
    return fig, ax
```

```python
def calc_ss(data, ss_type="two factor", effect="column"):
    data = np.array(data, dtype=np.int16)
    cell_samplesize = np.size(data, axis=2)
    no_column = np.size(data, axis=0)
    no_row = data.size / (no_column * cell_samplesize)
    
    if ss_type == "two factor":
        cell_totals = data.sum(axis=2)

        ss_between = np.sum(data.sum(axis=2)**2 / cell_samplesize) - data.sum()**2 / data.size
        ss_column = np.sum(cell_totals.sum(axis=1)**2 / (no_row*cell_samplesize)) - data.sum()**2 / data.size
        row_totals = list()
        for idx in range(int(no_row)):
            row_totals.append(np.sum(cell_totals[:, idx]))

        row_totals = np.array(row_totals)
        ss_row = np.sum(row_totals**2 / (no_column*cell_samplesize)) - data.sum()**2 / data.size

        ss_interaction = ss_between - (ss_column + ss_row)

        ss_within = np.sum(data**2) - np.sum(cell_totals**2 / cell_samplesize)

        return pd.Series({"ss_column": ss_column,
                          "ss_row": ss_row, 
                          "ss_interaction": ss_interaction,
                          "ss_within": ss_within})
    
    if ss_type == "simple effect":
        cell_totals = data.sum(axis=2).T if (effect == "column") else data.sum(axis=2)
        total_samplesize = cell_samplesize * (no_row if (effect == "column") else no_column)
        
        se_totals = np.sum(cell_totals**2, axis=0) / cell_samplesize - cell_totals.sum(axis=0)**2 / total_samplesize
        
        return pd.Series(se_totals, index=[f"level_{idx}" for idx in range(no_column if (effect == "column") else no_row)])
```

```python
def calc_df(data):
    data = np.array(data)
    
    df_column = np.size(data, axis=0) - 1
    df_row = data.size / (np.size(data, axis=0) * np.size(data, axis=2)) - 1
    df_interaction = df_column * df_row
    df_within = data.size - data.size / np.size(data, axis=2)
    
    return pd.Series({"df_column": df_column,
                      "df_row": df_row,
                      "df_interaction": df_interaction,
                      "df_within": df_within})
```

```python
def calc_ms(ss, df, ms_type="two factor"):
    if ms_type == "two factor":
        return pd.Series(np.round(ss / df, decimals=2),
                         index=["ms_column", "ms_row", "ms_interaction", "ms_within"])
    
    elif ms_type == "simple effect":
        return pd.Series(np.round(ss / df, decimals=2),
                         index=[f"ms_{idx}" for idx in range(len(df))])
```

```python
def calc_fscore(ms, score_type="two factor"):
    if score_type == "two factor":
        return pd.Series(np.round(ms[:"ms_interaction"].values / ms["ms_within"], decimals=2),
                         index=["fscore_column", "fscore_row", "fscore_interaction"])
    
    elif score_type == "simple effect":
        return pd.Series(np.round(ms / ms_within, decimals=2),
                         index=[f"fscore_level{idx}" for idx in range(ms.size)])
```

```python
def calc_partialetasquared(ss):
    col = ss["ss_column"]
    row = ss["ss_row"]
    interaction = ss["ss_interaction"]
    within = ss["ss_within"]
    
    return pd.Series(dict(etasq_column = col / (col + within),
                          etasq_row = row / (row + within),
                          etasq_interaction = interaction / (interaction + within)))
```

```python
def mean_diff(means):
    diff_list = list()
    
    for mean in means:
        diff = means - mean
        diff[diff <= 0] = np.nan
        diff_list.append(diff)
        
    return pd.DataFrame(data=diff_list, index=means, columns=means).round(decimals=2)
```
