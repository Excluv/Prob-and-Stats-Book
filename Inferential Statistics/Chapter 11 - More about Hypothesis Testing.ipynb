### **11.1** WHY HYPOTHESIS TEST?

- The observed sample mean is first of all, not a census, thus serves as only one of many possible means within the sampling distribution. Even though the probability is small, it may still lead to false conclusion when used as a generalization to the underlying population.
- Accordingly, the variability among samples means must be considered when we attempt to decide whether the observed difference between the sample mean (used as a generalization to the underlying population) and the population (against which, the underlying population is compared) mean is real or merely transitory (that the sample mean of another sample, different from the first produces a different conclusion).

#### **Importance of the Standard Error**

- To evaluate the effect of chance, there arises the need for the sampling distribution, that is, the concept of the sample means for all possible random samples.
- The difference between observed mean and population mean cannot be taken as a face value, rather, we need to convert the number into a multiple of deviation unit, which, in the case of sampling distribution, is the standard error.
- Before generalizing beyond the existing data, we must always measure the effect of chance; that is, we must obtain a value for the standard error.

#### **Possibility of Incorrect Decisions**

- Even with that being said, we can never know definitely if the decision is true or incorrect.
- The incorrect conclusions obtained from the hypothesis test may come in different forms, such as:
    + False positive: that the $H_0$ is incorrect, nonetheless, reported as true.
    + False negative: that the $H_0$ is actual correct, but reported as false.

#### **Minimizing Incorrect Decisions**

- Though it's impossible to calculate an exact number, based on the distribution of probabilities, most decisions will be correct - assuming the null hypothesis are either true or seriously false.

### **11.2** STRONG OR WEAK DECISIONS

- There are subtle but important differences in the interpretation of decisions to retain or to reject the $H_0$

#### **Retaining $H_0$ Is a <i>Weak</i> Decision**

- To prove that $H_0$ is true, one will need to have an actual census of all available data, so that calculated can be made in order to produce an exact conclusion.
- Nonetheless, oftentimes, the available datasets are only random samples, if there exists even one sample that disapproves the decision, then it will break easily. On the other hand, the given one sample that proves the opposition to the decision easily makes $H_0$ a false statement, therefore, the retention of $H_0$ is considered <i>weak</i>.
- Thus, the decision only implies that $H_0$ is probably not false, and <i>could</i> be true.

#### **Rejecting $H_0$ Is a <i>Strong</i> Decision**

- As mentioned above, if there is an evidence that $H_0$ should be rejected, then the decision is <i>strong</i> and logically firm.

#### **Why the Research Hypothesis Isn't Tested Directly**

- <b>Lacks necessary precision:</b> Testing the research hypothesis requires an exact number, whereas the available data are only suitable for the null hypothesis.
- <b>Supported by a strong decision to reject:</b> 
    + Because the research hypothesis is identified with the alternative hypothesis - the decision to reject the null hypothesis, should it be made, will provide a strong support for the research hypothesis, while the decision to retain the null hypothesis, should it be made, will provide, at most, a weak support for the null hypothesis.

### **11.3** ONE-TAILED TEST AND TWO-TAILED TESTS

- Generally, $H_1$ is the complement of the null hypothesis, $H_0$. 
- When converted to z scores, they lie beyond the critical values and account for 5% of the total distribution, which is also the <b>Level of Significance</b>.

#### **Two-Tailed Test**

- A hypothesis test, in which the rejection regions of $H_0$ are associated with both tails of the hypothesized sampling distribution, often referred to as the <b>Nondirectional Test</b>.
- Representation: $H_1: \mu \neq \mu_{pop}$.
- Critical values: $\pm$1.95.

#### **One-Tailed Test (Lower Tail Critical)**

- A hypothesis test, in which the rejection region of $H_0$ is associated with only the lower tail of the hypothesized sampling distribution, often referred to as the <b>Directional Test</b>.
- Representation: $H_1: \mu \lt \mu_{pop}$.
- Critical value: $-$1.65.

#### **Extra Sensitivity of One-Tailed Test**

- This kind of test is more sensitive to any drop in the observed mean, since the increase in critical value makes the region of rejection for $H_0$ bigger.
- Therefore, the probability of $H_0$ being rejected is more likely to occur than in nondirectional test.

#### **One-Tailed Test (Upper Tail Critical)**

- This kind of test mirros its counterpart, the One-Tailed Test (Lower Tail Critical), with its representation being: $H_1: \mu \gt \mu_{pop}$.
- Critical value: 1.65.

#### **Summary: The Three Types of Hypothesis Test**

#### **One or Two Tails?**

- Before a hypothesis test, if there is a concern that the true population mean differs from the hypothesized population only in a particular direction, use the appropriate one-tailed or directional test for extra sensitivity. Otherwise, use the more customary two-tailed or nondirectional test.

#### **New Null Hypothesis for One-Tailed Tests**

- Whenever nondirectional test is employed, write $H_0$ to include values of the population mean in the direction of no concern - even though the single number in the complete $H_0$ identified by the equality sign is the one value about which the hypothesized sampling distribution is centered about and, therefore, the one value actually used in the hypothesis test.

### **11.4** CHOOSING A LEVEL OF SIGNIFICANCE ($\alpha$)

- The level of significance indicates how rare the observed z score should be in order to reject $H_0$.
- It is not an absolute risk-free measurement, but in fact, carries the inherent risk, that the true $H_0$ is rejected. However, the risk is relatively small, since the probability of $H_0$ rejection, regardless of it being true or incorrect, is always equal to the level of significance.

#### **Which Level of Significance?**

- Except for the customary level of 0.05, the level of significance varies in different investigations, and is up to the investigators to decide, depends on the particular argument one makes.
- Another common level of $\alpha$ is 0.01, which has the z scores of $\pm$2.33.
- Note: unless there are obvious reasons to selecting a different level of significance, always use the customary 0.05.
- Summar: Common levels of significance <br>

### **11.7** IF $H_0$ IS REALLY TRUE

- Assume that $H_0$ is true, then the hypothesized distribution can be considered as the <i>true</i> sampling distribution, from which the observed mean originates (because the mean is centered about a single number, when variability is accounted).
- Even in that scenario, there's always a small probability that $H_0$ is wrongly rejected, because of the rejection region (which equals the level of significance).
- Graphical representation of the sampling distribution of a true $H_0$: <br>


#### **Probability of a Type I Error (Alpha)**

- The probability of type I error, that is, the rejection of a true null hypothesis is always equal to Alpha ($\alpha$) - the level of significance.

#### **Probability of a Correct Decision**

- Because the area under a normal distribution totals as 1, the probability (or the density) of $H_0$ retention equals 1 - $\alpha$. 

#### **Reducing the Probability of a Type I Error**

- To reduce the effect of type I errors, the investigator needs to lower the level of significance, for example, from 0.05 to 0.01 or even to 0.001.

#### **True $H_0$ Is Usually Retained**

- Because the probability of true $H_0$ retention is much larger than rejection, theoretically, $H_0$ will be more likely to be retained in the long run.

### **11.8** IF $H_0$ IS REALLY FALSE BECAUSE OF A LARGE EFFECT

- Definition:
    + Effect: Any difference between a true and a hypothesized population mean.

- In the case of $H_0$ being seriously false, then the observed mean must have originated from another distribution, rather than the hypothesized distribution that is being used to be compared against.
- An example of a true sampling distribution mean differs from the hypothesized distribution mean by a large amount: <br>


#### **Hypothesized Sampling Distribution**

- Definition: Centered about the hypothesized population mean, this distribution is used to generate the decision rule.

#### **<i>True</i> Sampling Distribution**

- Definition: Centered about the true population mean, this distribution produces the one observed mean (or z).

#### **Low Probability of a Type II Error for a Large Effect**

- The probabiltiy of retaining a type II error is denoted as $\beta$ (Beta) and equals to the proportion of the true population distribution that is overlapped with the hypothesized distribution and has the z score being equal to or less than 1.65 (only in the case of the upper critical test).

#### **High Probability of a Correct Decision for a Large Effect**

- The probability of $H_0$ being rejected will be more likely than the regular hypothesis test, which is equal to 1 - $\beta$.

### **11.9** IF $H_0$ REALLY IS FALSE BECAUSE OF A SMALL EFFECT

- The projected hypothesis test does not fare nearly as well if $H_0$ really is false because the difference between the true population mean and the hypothesized population mean is small.
- Similar to the case mentioned in the previous section, the probability for the retention of the false $H_0$ still accounts for the proportion that is overlapped with an area of the hypothesized distribution, which has its z score being equal to or lower than 1.65.
- An example of a small effect: <br>


#### **Low Probability of a Correct Decision for a Small Effect**

- Because the entire true population distribution is shifted toward the retention area, it increases the probability of producing the one observed mean that can result in a type II error decision, thus, lowers the probability of resulting a correct decision (relative to a false $H_0$ because of a large effect) (1 - $\beta$, the higher Beta the lower probability for correct decisions).

#### **Rejection of False $H_0$ Depends on Size of Effect**

- Summary: If $H_0$ really is false, then the probability of type II error ($\beta$) and the probability of a correct decision (1 - $\beta$) depend on the size of effect, that is, the difference between the true population mean and the hypothesized distribution mean. The smaller the size of effect, the higher probability of type II error and the lower probability of a correct decision.

### **11.10** INFLUENCE OF SAMPLE SIZE

- To increase the probability of detecting a false $H_0$, increase the sample size.
- As the sample size increases, the standard error decreases, thus narrows the range of the observable sample means and also the overlapped area  that lies beyond the critical value.
- An example of a small effect with larger sample size (relative to the example shown in the previous section): <br>


#### **Consequences of Reducing Standard Error**

- It shrinks the retention region back toward the hypothesized distribution mean.
- It shrinks the entire true population distribution back toward the true population mean.

#### **Samples Can Be Too Large**

- Extra large sample size can produce an extra sensitive hypothesis test that detects even the slightest, unimportant effect, which in fact, can be dismissed due to its lack of influence.

#### **Samples Can Be Too Small**

- To the contrary, a too small sample size with a large standard error will be more likely to miss even a large effect produced by an experiment.

#### **Neither Too Large or Too Small**

- For the purposes of most investigators, a sample size of hundreds is excessively large and one of less than five is unduly small.
- There remains, of course, considerable latitude for sample size selection between these rough extremities. Statistics supplies investigators with charts, often referred to as <i>power curves</i>, to help select the appropriate sample size for a particular experiment.

### **11.11** POWER AND SAMPLE SIZE

- Definition:
    + Power: The probability (1 - $\beta$) of detecting a particular effect when the null hypothesis ($H_0$) is false, which is the complement of the probability of the retention of a false $H_0$.

- As mentioned above, the power depends entirely on the sample size, of which, the selection should reflect:
    + What constitutes the smallest yet important effect: the size of effect should be determined based on some particular standard or, due to Jacob Cohen's widely adopted guidelines, equals to a half of the population standard deviation, which is commonly called a "medium" effect size. 
    + A reasonable degree of power for detecting that effect: using some calculations, the investigators will be able to select the appropriate degree of power according to the selected sample size.

#### **Power Curves**

- Definition:
    + Power Curve: A curve that shows how the likelihood of detected any possible effect - ranging from very small to very large - varies for a fixed sample size.

- An example of the power curves: <br>


- When it is not favorable to select a larger sample size, there are a few techniques to increase the probability of detecting a false $H_0$ as follows:
    + Enlarge the smallest important effect (preferably by only small increments): a larger threshold for the important effect will always produce a higher degree of power.
    + Lower the degree of power (preferably not much below 0.80): a resonable lower degree of power can be resulted from a smaller sample size.
    + Increase the level of significance (preferably not above 0.10): a larger rejection area will decrease the probability of the retention of a false $H_0$.
    + Select a one-tailed test rather than a two-tailed test, given that it is appropriate.
    + Combine various techniques as mentioned above.

- Note:
    + The use of the power curves represents a distinct improvement over the arbitrary selection of sample size, for power curves help identify a sample size that, being neither unduly small nor excessively large, produces a hypothesis test with the proper sensitivity.

#### **Power Analysis of Studies by Others**

- If you suspect the reported failure to reject $H_0$ might have been caused by an unduly small sample size, power curves can be consulted retroactively to evaluate the adequacy of the result by examining the degree of power given from the sample size applied in the investigation. 

#### **Need Not Predict True Effect Size**

- The effect size in the power analysis needs not be perfect, and rather can be a random guess from a series of numbers. However, it is always favorable to have some standard, against which, the desired effect size is compared, so that the effect sizes that merit detection can be specified.
- If the true effect size actually exceeds the hypothesized important effect, then the power is really higher than that in the hypothesis, therefore, more favorable to the investigator to reject the false $H_0$.
- On the other hand, if the true effect is lower than the hypothesized important effect, then it is more likely to retain the false $H_0$, but the missed effect, was it to be detected, would be dismissed anyway.
