### **Chapter 14:** t Test for Two Independent Samples

#### **Purpose:**

- As in t Test for One Sample, this test is designed to evaluate whether the observed mean difference between two set of samples, each of which receives a different independent variable, is real or merely transitory when the true population distributions are unknown.

#### **Terminology:**

- Two independent samples: the groups of subjects that are examined under two distinct independent variables.
- Sampling distribution of $\overline{X}_1 - \overline{X}_2$: a hypothesized sampling distribution of the differences between sample means.
- Pooled variance estimate ($s^2_p$): an estimate of the population variance (assumed to be the same for both populations).
- Estimated standard error ($s_{\overline{X}_1 - \overline{X}_2}$): the standard deviation of the sampling distribution for the differences between means.
- Effect: the real difference between true population means.
- Confidence intervals for $\mu_1 - \mu_2$: an interval that is likely to include the true population mean.
- Standardized effect estimate, Cohen's $d$: the estimated effect size that is stable upon the changes in sample size.
- Meta-analysis: a set of data-collection and statistical procedures designed to summarize the various effects reported by groups of similar studies.
- File drawer effect: a false positive caused by an inflated type I error attributable to reports of nonsignificant findings not being publicized. 

#### **Core Principle:**

- This t test considers the direct difference between the samples that receive different independent variables after it has been measured against the variability within group.

- The t test is based on a family of t distributions, each of which is associated with a different degrees of freedom. After a series of calculations, we will obtain a few indicators, which are:
    + The t score: represents the relative location of the observed mean difference about the hypothesized sampling distribution mean.
    + The approximate (or exact) $p$-value: the proportion or probability for such a mean difference or larger values to occur.
- Even when the result is determined to be statistically significant, this conclusion provide very little information about the hypothesis, so that it is suggested to conduct a couple of estimations to further materialize the test result and get addtional evidences for an expanded coverage of the decision, those estimations include:
    + Confidence interval: represents a range that is likely to include the true population mean difference.
    + Stadardized estimate effect size Cohen's $d$: an indicator of the probable effect size.
- What we obtain from the t test is actually not a concrete evidence, rather a statistically supported foundation to make a decision about the present hypothesis. Therefore, replication or further reference to other similar studies is recommended in order to fortify our understanding of the hypothesis or the given study's problem.
- Assumptions: of the current t test
    + The samples originate from normally distributed populations.
    + These populations have equal variances.

### **Chapter 15:** t Test for Two Related Samples

#### **Purpose**

- The t test for two related samples measures the difference between data obtained from a same group or two matching groups (on pair-wise basis, that is, each subject of a group is associated with one and only one other subject in another group) of samples when they receive different independent variables and the true population distribution is unknown.

#### **Terminology:**

- Difference score ($D$): the differences between pairs of scores, each of which is taken from a different group or the same group but under different independent variable.
- Repeated measures: the measures that are taken repeatedly on a same group or two matching groups.
- Two related samples: the samples that either consist of the same subjects or of different subjects who possess matching characteristics.
- Counterbalancing: the procedure of reversing the order of conditions for equal numbers of all subjects.
- Population correlation coefficient ($p$): the true or hypothesized (if unknown) correlation coeffcient of the variables in a population.

#### **Core Principle:**

- Similar to that in other t tests, this test's main purpose is to identify whether the difference of observed means between samples is real or merely transitory. It is done by comparing the direct difference between observed means against the variability within subjects (or the so call error varability) that occurs because of the obvious discrepancy in nature. However, the variability because of individual differences in this t test is minimized or smaller than that of two independent samples since the subjects share very similar characteristics based on some selection in the beginning.

- Some complications that need considerations:
    + Should there exist any lingering effect due to a particular treatment, sufficient time must elapse in order to eliminate any potential uncontrolled factors.
    + The order of treatments can also cause the subjects to have some certain fallacies, therefore, counterbalancing or similar methods must take place to ensure that there's no interference in the process of current study.

### **Chapter 16:** Analysis of Variance (One Factor)

#### **Purpose:**

- The one factor ANOVA is designed to measure the differences between samples that receive the same independent variable but on different levels.

#### **Terminology:**

- Analysis of variance (ANOVA): an overall test of the null hypothesis for more than two population means.
- Treatment effect: the existence of at least one difference between the population means.
- Variability within group: the variability among scores of subjects who, being in the same group, receive the same experimental treatment.
- Random error: another term for the variability within group (it is random because of the obvious discrepancies between subjects that occurs by nature).
- Variability between groups: the variability among scores of subjects who, being in different groups, receive different experimental treatment.
- Mean square (MS): a variance estimate
- F ratio: ?
- Squared curvilinear correlation ($\eta^2$): the proportion of variance in the dependent variable that can be explained by or attributed to the independent variable.
- Multiple Comparison: all possible comparisons when two or more population means are involved.
- Turkey's HSD test: a multiple comparison test for which the cumulative probability of a type I error never exceeds the specified level of significance.

#### **Core Principle:**

- By measuring the two sources of variability through various calculation procedures, we obtain a number - the F ratio, which represents essentially how many times the variability within group (reflects the random error) is compared to the variability between groups (reflects the effect caused by the difference in treatments). When F is examined based on the degrees of freedom in the two sources of variability and proves to be sufficiently large at some critical level, then we conclude that there's a real difference caused by the treatments. Otherwise, it is considered transitory due to chance (because variability is always accounted for).
- The F test is always a (positive) directional test due to the involvement of squared deviation scores.
- Assumptions:
    + The samples are of equal size.
    + The samples originate from normally distributed populations and have equal variances.
    + The sample size is sufficiently large (greater than about 10).

#### **Terminology:**

- Partial squared curvilinear correlation ($\eta^2_p$): the proportion of explained variance in the dependent variable by the independent variable after one or more sources of variablity have been eliminated from the total variance.

#### **Core Principle:**

### **Chapter 18:** Analysis of Variance (Two Factors)

#### **Purpose:**

- This type of ANOVA test implements two factors (independent variables), within each are multiple levels of treatment, in order to measure the effects of not just the two given factors but also to observe to interaction between them.

#### **Terminology:**

- Interaction: the product of the inconsistent simple effects.
- Main effect: the effect of a single factor when any other factor is ignored.
- Simple effect: the effect of one factor (every level of that factor is inclusive) on the dependent variable at a single level of another factor.
- $F_{se}$ test for simple effect: a test for simple effects.

#### **Core Principle:**

- When two factors are involved in an experiment, there are multiple possible outcomes, based on how many levels of treatment each factor possesses. Therefore, in order to interpret the result correctly, an investigator has to first evaluate the significance of the interaction between independent variables before moving on to examine the effects of each factor.
- Two-factor ANOVA still measures obtained data the same way as in other F tests, with an additional information about the interaction between factors (independent variables). However, this type of F test is always more efficient than that of one-factor due to the more concise implementation and it being more informative. 
- Assumptions: besides the assumptions of other similar F tests
